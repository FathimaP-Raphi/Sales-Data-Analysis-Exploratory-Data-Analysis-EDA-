# SALES & E-COMMERCE EXPLORATORY DATA ANALYSIS

Project Objective

---



This project performs a comprehensive Exploratory Data Analysis (EDA) on a Retail sales dataset to uncover meaningful business insights. The analysis aims to:

- Understand **sales patterns** across branches, cities, product lines, and customer segments  
- Identify **top-performing products** and **revenue drivers**

Dataset Description

---



The dataset contains transaction-level sales data from a supermarket chain operating across multiple branches, where each row represents a single purchase.

Key Features:
Transaction Details: Invoice ID, Date, Time
Customer Information: Gender, Customer Type (Member/Normal)
Location Data: Branch, City
Product Information: Product Line (category)
Sales Metrics: Unit Price, Quantity, Total, Tax, Cost of Goods Sold (COGS)
Payment Method: Cash, Credit Card, E-wallet
Customer Feedback: Rating (1–10 scale)


**Problem Statement**

How can the supermarket chain optimise its sales strategy, product offerings and customer experience across its three branches to maximise revenue and customer satisfaction


---


Specific questions this analysis addresses:

1. Which **branch and city** generates the highest revenue?  
2. Which **product lines** are most and least profitable?  
3. How do **gender and customer type** influence purchasing behaviour?  
4. Is there a relationship between **customer rating and total spend**?  
5. Which **payment methods** are most preferred, and does preference vary by branch?

**Fix Data Types — Date & Time**


```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

print("Date dtype after conversion:", df['Date'].dtype)
print("Date range:", df['Date'].min().date(), "-", df['Date'].max().date())
```

**Derived Features**
1) Extracting Temporal features Name of the month and day of the week from Date and Time column.


```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['Month name'] = df['Date'].dt.strftime('%B')
df['Day of week'] = df['Date'].dt.day_name()
df['Hour'] = pd.to_datetime(df['Time'].astype(str), format='%H:%M:%S').dt.hour
```

2) Revenue per unit


```python
df['Revenue per unit'] = df['Total'] / df['Quantity']
```


3. Extracting Day type whether day of the week falls on week day or week end.

```python
df['Day type'] = df['Day of week'].apply(
    lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Weekday'
)
```

4) Customer Satisfaction Segmentation

---



To improve business insight readability, the Customer Satisfaction Level feature is added by binning the continuous **Rating** data.

* **0 - 5:** Classified as **Low** (Needs attention/At-risk customers)
* **5 - 8:** Classified as **Medium** (Neutral/Satisfied customers)
* **8 - 10:** Classified as **High** (Highly satisfied/Brand promoters)

Transforming continuous ratings into distinct categorical bins makes it much easier to build aggregated visualizations and perform customer segmentation analysis.

```python
df['Customer Satisfaction Level'] = pd.cut( df['Rating'], bins=[0, 5, 8, 10], labels=['Low', 'Medium', 'High'], include_lowest=True)
```

9. **Daily Sales Trend**

---



```python
daily = df.groupby('Date')['Total'].sum().reset_index()
fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(daily['Date'], daily['Total'], color='steelblue', linewidth=1.5, alpha=0.8)
ax.fill_between(daily['Date'], daily['Total'], alpha=0.15, color='steelblue')
ax.set_title('Daily Total Sales Trend (Jan – Mar 2019)', fontweight='bold', fontsize=13)
ax.set_xlabel('Date')
ax.set_ylabel('Total Revenue (USD)')
# Rolling average
daily['Rolling_7'] = daily['Total'].rolling(7).mean()
ax.plot(daily['Date'], daily['Rolling_7'], color='red', linewidth=2,
        linestyle='--', label='7-day Rolling Average')
ax.legend()
plt.tight_layout()
plt.show()
print("Insight: Sales exhibit fluctuation with no strong single-peak trend,\n"
      "though the rolling average reveals a slight mid-period dip in February.")
```

10. **Sales by Hour of Day & Day of Week (Heatmap)**

---



```python
hour_day = df.pivot_table(values='Total', index='Hour', columns='Day of week', aggfunc='sum')

# Reorder days
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
hour_day = hour_day.reindex(columns=[d for d in day_order if d in hour_day.columns])

plt.figure(figsize=(8, 6))
sns.heatmap(hour_day, cmap='YlOrRd', linewidths=0.4, annot=True, fmt='.0f',cbar_kws={'label': 'Revenue (USD)'})
plt.title('Temporal Sales Patterns: Revenue Heatmap — Hour × Day', fontsize=10, fontweight='bold', pad=15)
plt.xlabel('Day of Week', fontsize=11, labelpad=10)
plt.ylabel('Hour of Day', fontsize=11, labelpad=10)
plt.tight_layout()
plt.show()
print("Insight: Peak shopping hours are 10–11 AM, 1 PM, and 7 PM.\n"
      "Saturdays and Tuesdays see the highest activity.")
```

**Multivariate Analysis**

11. **Monthly Revenue by Branch (Line Chart**)

```python
# Keep only the three main months present in dataset
month_branch = df.groupby(['Month', 'Branch'])['Total'].sum().reset_index()
month_name_map = {1: 'January', 2: 'February', 3: 'March'}
month_branch = month_branch[month_branch['Month'].isin([1, 2, 3])].copy()
month_branch['Month_label'] = month_branch['Month'].map(month_name_map)
month_branch = month_branch.sort_values('Month')

fig, ax = plt.subplots(figsize=(8, 5))
for branch, grp in month_branch.groupby('Branch'):
    ax.plot(grp['Month_label'], grp['Total'], marker='o', linewidth=2, label=f'Branch {branch}')
ax.set_title('Monthly Revenue Trend by Branch (Jan–Mar 2019)', fontweight='bold', fontsize=13)
ax.set_xlabel('Month')
ax.set_ylabel('Total Revenue (USD)')
ax.legend(title='Branch')
plt.tight_layout()
plt.show()
print("Insight: January is the strongest month for all branches.\n"
      "Branch C (Naypyitaw)and Branch A(Yangon) shows recovery in March while Branch B shows a dip.")
```

13. **Correlation Heatmap**

---



```python
num_df = df[['Unit price', 'Quantity', 'Tax 5%', 'Total', 'cogs',
             'gross income', 'Rating', 'Hour', 'Revenue per unit']].copy()

corr = num_df.corr()
fig, ax = plt.subplots(figsize=(11, 8))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, linewidths=0.5, ax=ax, square=True,
            cbar_kws={'shrink': 0.8})
ax.set_title('Correlation Matrix of Numerical Features', fontweight='bold', fontsize=13)
plt.tight_layout()
plt.show()
print("Insight: Total, Tax 5%, COGS, and Gross Income are perfectly correlated\n"
      "(mathematically derived). Rating shows near-zero correlation with spend \n"
      "customers rate independently of how much they spent.")
```

**KEY INSIGHTS:**

---
**Insight 1 — Balanced Branch Performance**
All three branches (Yangon, Naypyitaw, Mandalay) contribute nearly **equal revenue (~33% each)**.  
This indicates a well-distributed customer base, but also that **no single branch dominates**, making branch-specific strategies critical.

---

**Insight 2 — Food & Beverages Leads Revenue; Fashion Accessories Leads Transactions**
**Food and beverages** generate the highest total revenue, while **fashion accessories** records  
the most transactions. This suggests Food & Beverages has a **higher average basket size** —  
an opportunity to expand offerings or run combo promotions.

---

**Insight 3 — Peak Shopping Hours: 10–11 AM, 1 PM, 7 PM**
Transaction volumes are highest in the **late morning, afternoon, and early evening**.  
Staffing, promotions, and stock replenishment should be concentrated around these windows  
to maximise throughput and customer satisfaction.

---

**Insight 4 — Customer Rating is Independent of Spending**
The correlation between **Rating** and **Total** is near zero (≈ 0.0).  
Customers do **not rate higher because they spent more** — satisfaction is driven by service quality and product experience, not transaction value.  
This means even low-spend customers can be brand ambassadors.

---

**Insight 5 — E-Wallet is the Dominant Payment Method**
E-wallets account for the largest share of transactions (~34%), surpassing cash and credit cards.  
Supermarkets should incentivise digital payments further (cashback, loyalty points)  
to reduce cash-handling costs and accelerate checkout speed.
